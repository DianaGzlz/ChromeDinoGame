### Hablemos de NEAT 

- ¿Qué es NEAT?

NEAT, acrónimo de NeuroEvolution of Augmenting Topologies (Neuroevolución de topologías en aumento), es un algoritmo evolutivo utilizado en el campo de la inteligencia artificial. A diferencia de las redes neuronales tradicionales, donde se ajustan los pesos y sesgos, NEAT se centra en modificar la topología de la red neuronal. Esto significa que durante el entrenamiento, la estructura de la red neuronal cambia añadiendo y eliminando conexiones y nodos.

- ¿Cómo funciona NEAT?

NEAT funciona de manera similar a la evolución natural. Se inicia con una población de redes neuronales aleatorias y luego se itera a través de las siguientes generaciones:

1. Evaluación: Cada red neuronal se evalúa en su capacidad para realizar una tarea específica. Se les asigna una puntuación de aptitud en función de su rendimiento.
2. Selección: Las redes neuronales con mayor puntuación de aptitud tienen más probabilidades de ser seleccionadas para la reproducción.
3. Reproducción: Las redes neuronales seleccionadas se "reproducen" para crear nuevas redes neuronales. Esto se hace mediante la combinación de sus estructuras y la introducción de mutaciones aleatorias.
4. Repetición: Los pasos 1 a 3 se repiten durante varias generaciones. A medida que las generaciones progresan, las redes neuronales se vuelven cada vez más aptas para realizar la tarea objetivo.

**Ventajas de NEAT:**

* Flexibilidad: NEAT permite que la estructura de la red neuronal se adapte y evolucione durante el entrenamiento, lo que la hace más adecuada para problemas complejos.
* Versatilidad: NEAT puede aplicarse a una amplia gama de problemas de aprendizaje automático, incluyendo tareas de clasificación, predicción y control.
* Eficiencia: NEAT puede ejecutarse en CPUs convencionales, sin necesidad de hardware especializado.

**Desventajas de NEAT:**

* Tiempo de entrenamiento: El proceso de entrenamiento de NEAT puede ser lento, debido a la naturaleza evolutiva del algoritmo.
* Complejidad: NEAT es un algoritmo complejo que puede ser difícil de implementar y ajustar.

**Aplicaciones de NEAT**

NEAT se ha utilizado con éxito en una variedad de aplicaciones, incluyendo:

* Juegos: NEAT se ha utilizado para entrenar redes neuronales que pueden jugar juegos como Super Mario Bros. y Tetris a un nivel superhumano.
* Robótica: NEAT se ha utilizado para controlar robots que pueden caminar, correr y navegar por entornos complejos.
* Finanzas: NEAT se ha utilizado para desarrollar modelos de predicción de mercado.
En general, NEAT es una herramienta poderosa para el diseño de redes neuronales eficientes y adaptables. Es una opción viable para una amplia gama de problemas de aprendizaje automático, especialmente aquellos que requieren estructuras de red neuronal complejas.

```python
#agregar IA al juego
import pygame
import os
import random
import math #nuevo
import neat #nuevo 
import  sys

import pygame.image
import pygame.time



#Iniciar juego
pygame.init()

#constantes globales
SCREEN_HEIGHT = 600
SCREEN_WIDTH = 1100
SCREEN = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

RUNNING = [pygame.image.load(os.path.join("Assets/Dino", "DinoRun1.png")),
           pygame.image.load(os.path.join("Assets/Dino", "DinoRun2.png"))]

JUMPING = pygame.image.load(os.path.join("Assets/Dino", "DinoJump.png"))

SMALL_CACTUS = [pygame.image.load(os.path.join("Assets/Cactus", "SmallCactus1.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "SmallCactus2.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "SmallCactus3.png"))]
LARGE_CACTUS = [pygame.image.load(os.path.join("Assets/Cactus", "LargeCactus1.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "LargeCactus2.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "LargeCactus3.png"))]



BG = pygame.image.load(os.path.join("Assets/Other", "Track.png"))

FONT = pygame.font.Font('freesansbold.ttf', 20)

class Dinosaur:
    X_POS = 80
    Y_POS = 310
    JUMP_VEL = 8.5

    def __init__(self, img=RUNNING[0]):
        self.image = img #crear el dinosaurio
        self.dino_run = True #estado inicial
        self.dino_jump = False
        self.jump_vel = self.JUMP_VEL 
        self.rect = pygame.Rect(self.X_POS, self.Y_POS, img.get_width(), img.get_height()) #crea un rectangulo alrededor del dinosaurio, iniciando en esquina superior izquierda
        self.step_index = 0

    def update(self):
        if self.dino_run:
            self.run()
        if self.dino_jump:
            self.jump()
        if self.step_index >= 10:
            self.step_index = 0

    def jump(self):
        self.image = JUMPING
        if self.dino_jump:
            self.rect.y -= self.jump_vel * 4
            self.jump_vel -= 0.8
        if self.jump_vel <= -self.JUMP_VEL:
            self.dino_jump = False
            self.dino_run = True
            self.jump_vel = self.JUMP_VEL

    def run(self):
        self.image = RUNNING[self.step_index // 5]
        self.rect.x = self.X_POS
        self.rect.y = self.Y_POS
        self.step_index += 1

    def draw(self, SCREEN): #CADA 5 CUADROS CAMBIA LA IMAGEN DEL DINOSAURIO CAMINANDO (PASO)
        SCREEN.blit(self.image, (self.rect.x, self.rect.y))

        
class Obstacle:
    def __init__(self, image, number_of_cacti):
        self.image = image
        self.type = number_of_cacti
        self.rect = self.image[self.type].get_rect()
        self.rect.x = SCREEN_WIDTH

    def update(self):
        self.rect.x -= game_speed
        if self.rect.x < -self.rect.width:
            obstacles.pop()

    def draw(self, SCREEN):
        SCREEN.blit(self.image[self.type], self.rect)


class SmallCactus(Obstacle):
    def __init__(self, image, number_of_cacti):
        super().__init__(image, number_of_cacti)
        self.rect.y = 325


class LargeCactus(Obstacle):
    def __init__(self, image, number_of_cacti):
        super().__init__(image, number_of_cacti)
        self.rect.y = 300


def remove(index): ######AUJSTE
    dinosaurs.pop(index)
    ge.pop(index)
    nets.pop(index)


#################NUEVO################
def distance(pos_a, pos_b):
    dx = pos_a[0]-pos_b[0]
    dy = pos_a[1]-pos_b[1]
    return math.sqrt(dx**2+dy**2)
########################################


def eval_genomes(genomes, config): #ajustes aqui, cambio nombre de main a "eval_genomes"
    global game_speed, x_pos_bg, y_pos_bg, obstacles, dinosaurs, ge, nets, points
    clock = pygame.time.Clock()
    points = 0

    obstacles = [] 
    dinosaurs = [] #quitamos el dinosaurio que agregamos previamente
    ge = [] #dicionarios, información de cada dinosaurio, sus notas y las conecciones de la red neuronal
    nets = [] #inicia la red neuronal por dinosaurio

    x_pos_bg = 0
    y_pos_bg = 380
    game_speed = 20

    #############NUEVO############
    for genome_id, genome in genomes:
        dinosaurs.append(Dinosaur())
        ge. append(genome)
        net = neat.nn.FeedForwardNetwork.create(genome, config)
        nets.append(net)
        genome.fitness = 0
    ####################################

    def score():
        global points, game_speed
        points += 1
        if points % 100 == 0:
            game_speed += 1
        text = FONT.render(f'Points: {str(points)}', True, (0, 0, 0))
        SCREEN.blit(text, (950, 50))


    def background():
        global x_pos_bg, y_pos_bg
        image_width = BG.get_width()
        SCREEN.blit(BG, (x_pos_bg, y_pos_bg))
        SCREEN.blit(BG, (image_width + x_pos_bg, y_pos_bg))
        if x_pos_bg <= -image_width:
            x_pos_bg = 0
        x_pos_bg -= game_speed



    run = True
    while run:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
        SCREEN.fill((255, 255, 255))

        for dinosaur in dinosaurs:  
            dinosaur.update()
            dinosaur.draw(SCREEN)

        if len(dinosaurs) == 0:
            break
        
        if len(obstacles) == 0:
            rand_int = random.randint(0,1)
            if rand_int == 0:
                obstacles.append(SmallCactus(SMALL_CACTUS, random.randint(0,2)))
            elif rand_int == 1:
                obstacles.append(LargeCactus(LARGE_CACTUS, random.randint(0,2)))

        for obstacle in obstacles:
            obstacle.draw(SCREEN)
            obstacle.update()
            for i, dinosaur in enumerate(dinosaurs):
                if dinosaur.rect.colliderect(obstacle.rect):
                    ge[i].fitness -= 1 ######nuevo#######
                    remove(i)
    

        # user_input = pygame.key.get_pressed() #ELIMINAR PORQUE LO HARA LA IA
        for i, dinosaur in enumerate(dinosaurs):
            ############### NUEVO ####################
            output = nets[i].activate((dinosaur.rect.y, #posición en y
                                       distance((dinosaur.rect.x, dinosaur.rect.y), #distamcia con respecto a obstaculo
                                                obstacle.rect.midtop)))
            if output[0] > 0.5 and dinosaur.rect.y == dinosaur.Y_POS: #modificamos aqui
                dinosaur.dino_jump = True
                dinosaur.dino_run = False
            ##############################################

        score()
        background()
        clock.tick(30) #cuadros por segundo
        pygame.display.update()

################### NUEVO#############
#setup NEAT

def run(config_path):
    config = neat.config.Config(
        neat.DefaultGenome,
        neat.DefaultReproduction,
        neat.DefaultSpeciesSet,
        neat.DefaultStagnation,
        config_path
    )

    pop = neat.Population(config)
    pop.run(eval_genomes, 50 ) #fitness function


if __name__ == '__main__':
    local_dir = os.path.dirname(__file__)
    config_path = os.path.join(local_dir, 'config.txt')
    run(config_path)
```

```python
#agregar info a juego con IA
import pygame
import os
import random
import math 
import neat 
import  sys

import pygame.draw
import pygame.image
import pygame.time



#Iniciar juego
pygame.init()

#constantes globales
SCREEN_HEIGHT = 600
SCREEN_WIDTH = 1100
SCREEN = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

RUNNING = [pygame.image.load(os.path.join("Assets/Dino", "DinoRun1.png")),
           pygame.image.load(os.path.join("Assets/Dino", "DinoRun2.png"))]

JUMPING = pygame.image.load(os.path.join("Assets/Dino", "DinoJump.png"))

SMALL_CACTUS = [pygame.image.load(os.path.join("Assets/Cactus", "SmallCactus1.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "SmallCactus2.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "SmallCactus3.png"))]
LARGE_CACTUS = [pygame.image.load(os.path.join("Assets/Cactus", "LargeCactus1.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "LargeCactus2.png")),
                pygame.image.load(os.path.join("Assets/Cactus", "LargeCactus3.png"))]



BG = pygame.image.load(os.path.join("Assets/Other", "Track.png"))

FONT = pygame.font.Font('freesansbold.ttf', 20)

class Dinosaur:
    X_POS = 80
    Y_POS = 310
    JUMP_VEL = 8.5

    def __init__(self, img=RUNNING[0]): #############AJUSTES
        self.image = img #crear el dinosaurio
        self.dino_run = True #estado inicial
        self.dino_jump = False
        self.jump_vel = self.JUMP_VEL 
        self.rect = pygame.Rect(self.X_POS, self.Y_POS, img.get_width(), img.get_height()) #crea un rectangulo alrededor del dinosaurio, iniciando en esquina superior izquierda
        self.color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) ######NUEVO
        self.step_index = 0

    def update(self):
        if self.dino_run:
            self.run()
        if self.dino_jump:
            self.jump()
        if self.step_index >= 10:
            self.step_index = 0

    def jump(self):
        self.image = JUMPING
        if self.dino_jump:
            self.rect.y -= self.jump_vel * 4
            self.jump_vel -= 0.8
        if self.jump_vel <= -self.JUMP_VEL:
            self.dino_jump = False
            self.dino_run = True
            self.jump_vel = self.JUMP_VEL

    def run(self):
        self.image = RUNNING[self.step_index // 5]
        self.rect.x = self.X_POS
        self.rect.y = self.Y_POS
        self.step_index += 1

    def draw(self, SCREEN): #CADA 5 CUADROS CAMBIA LA IMAGEN DEL DINOSAURIO CAMINANDO (PASO)
        SCREEN.blit(self.image, (self.rect.x, self.rect.y))
        ########## NUEVO ###########
        pygame.draw.rect(SCREEN, self.color, (self.rect.x, self.rect.y, self.rect.width, self.rect.height), 2) 
        for obstacle in obstacles:
            pygame.draw.line(SCREEN, self.color, (self.rect.x + 54, self.rect.y + 12), obstacle.rect.center, 2)

        
class Obstacle:
    def __init__(self, image, number_of_cacti):
        self.image = image
        self.type = number_of_cacti
        self.rect = self.image[self.type].get_rect()
        self.rect.x = SCREEN_WIDTH

    def update(self):
        self.rect.x -= game_speed
        if self.rect.x < -self.rect.width:
            obstacles.pop()

    def draw(self, SCREEN):
        SCREEN.blit(self.image[self.type], self.rect)


class SmallCactus(Obstacle):
    def __init__(self, image, number_of_cacti):
        super().__init__(image, number_of_cacti)
        self.rect.y = 325


class LargeCactus(Obstacle):
    def __init__(self, image, number_of_cacti):
        super().__init__(image, number_of_cacti)
        self.rect.y = 300


def remove(index):
    dinosaurs.pop(index)
    ge.pop(index)
    nets.pop(index)


def distance(pos_a, pos_b):
    dx = pos_a[0]-pos_b[0]
    dy = pos_a[1]-pos_b[1]
    return math.sqrt(dx**2+dy**2)


def eval_genomes(genomes, config): #ajustes aqui, cambio nombre de main a "eval_genomes"
    global game_speed, x_pos_bg, y_pos_bg, obstacles, dinosaurs, ge, nets, points
    clock = pygame.time.Clock()
    points = 0

    obstacles = [] 
    dinosaurs = [] #quitamos el dinosaurio que agregamos previamente
    ge = [] #dicionarios, información de cada dinosaurio, sus notas y las conecciones de la red neuronal
    nets = [] #inicia la red neuronal por dinosaurio

    x_pos_bg = 0
    y_pos_bg = 380
    game_speed = 20


    for genome_id, genome in genomes:
        dinosaurs.append(Dinosaur())
        ge. append(genome)
        net = neat.nn.FeedForwardNetwork.create(genome, config)
        nets.append(net)
        genome.fitness = 0

    def score():
        global points, game_speed
        points += 1
        if points % 100 == 0:
            game_speed += 1
        text = FONT.render(f'Points: {str(points)}', True, (0, 0, 0))
        SCREEN.blit(text, (950, 50))


    ################## NUEVO ##########################
    def statistics():
        global dinosaurs, game_speed, ge
        text_1 = FONT.render(f'Dinosaurs Alive: {str(len(dinosaurs))}', True, (0,0,0))
        text_2 = FONT.render(f'Generation: {pop.generation+1}', True, (0,0,0))
        text_3 = FONT.render(f'Game Speed: {str(game_speed)}', True, (0,0,0))

        SCREEN.blit(text_1, (50, 450))
        SCREEN.blit(text_2, (50, 480))
        SCREEN.blit(text_3, (50, 510))
    ##################################################



    def background():
        global x_pos_bg, y_pos_bg
        image_width = BG.get_width()
        SCREEN.blit(BG, (x_pos_bg, y_pos_bg))
        SCREEN.blit(BG, (image_width + x_pos_bg, y_pos_bg))
        if x_pos_bg <= -image_width:
            x_pos_bg = 0
        x_pos_bg -= game_speed



    run = True
    while run:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
        SCREEN.fill((255, 255, 255))

        for dinosaur in dinosaurs:  
            dinosaur.update()
            dinosaur.draw(SCREEN)

        if len(dinosaurs) == 0:
            break
        
        if len(obstacles) == 0:
            rand_int = random.randint(0,1)
            if rand_int == 0:
                obstacles.append(SmallCactus(SMALL_CACTUS, random.randint(0,2)))
            elif rand_int == 1:
                obstacles.append(LargeCactus(LARGE_CACTUS, random.randint(0,2)))

        for obstacle in obstacles:
            obstacle.draw(SCREEN)
            obstacle.update()
            for i, dinosaur in enumerate(dinosaurs):
                if dinosaur.rect.colliderect(obstacle.rect):
                    ge[i].fitness -= 1 
                    remove(i)
    

        for i, dinosaur in enumerate(dinosaurs):
            output = nets[i].activate((dinosaur.rect.y, #posición en y
                                       distance((dinosaur.rect.x, dinosaur.rect.y), #distamcia con respecto a obstaculo
                                                obstacle.rect.midtop)))
            if output[0] > 0.5 and dinosaur.rect.y == dinosaur.Y_POS: #modificamos aqui
                dinosaur.dino_jump = True
                dinosaur.dino_run = False


        statistics()  #NUEVO
        score()
        background()
        clock.tick(30) #cuadros por segundo
        pygame.display.update()

#setup NEAT

def run(config_path):
    global pop  ############## NUEVO############
    config = neat.config.Config(
        neat.DefaultGenome,
        neat.DefaultReproduction,
        neat.DefaultSpeciesSet,
        neat.DefaultStagnation,
        config_path
    )

    pop = neat.Population(config)
    pop.run(eval_genomes, 50 ) #fitness function


if __name__ == '__main__':
    local_dir = os.path.dirname(__file__)
    config_path = os.path.join(local_dir, 'config.txt')
    run(config_path)
```
